```python
# ==========================================
# Exercise 5: Clinical BMI Calculator
# ==========================================

while True:
    try:
        weight = float(input("Enter your weight in kilograms (kg): "))
        height = float(input("Enter your height in meters (m): "))
        
        bmi = weight / (height ** 2)
        
        if bmi < 18.5:
            status = "Underweight"
        elif 18.5 <= bmi <= 24.9:
            status = "Normal weight"
        else:
            status = "Overweight"
            
        print(f"\nYour BMI is {round(bmi, 1)}. Weight status: {status}")
        break # Exit loop if input is successful
    except ValueError:
        print("Invalid input! Please enter numbers only and do not leave blank.")


# ==========================================
# Exercise 6: Lecture Theatre Occupancy Check
# ==========================================

max_capacity = 120

while True:
    try:
        attendees = int(input("\nEnter the number of people expected to attend: "))
        
        if attendees <= max_capacity:
            remaining = max_capacity - attendees
            print(f"The event is legal. You have {remaining} available spots.")
        else:
            excess = attendees - max_capacity
            print(f"The event is illegal! {excess} people must be turned away.")
        break
    except ValueError:
        print("Invalid input! Please enter a whole number.")


# ==========================================
# Exercise 7: Pharmacy Inventory Report
# ==========================================

pharmacy_inventory = {
    'Analgesics': ['Paracetamol', 'Ibuprofen'],
    'Antibiotics': ['Amoxicillin', 'Doxycycline', 'Penicillin V'],
    'Antiseptics': ['Iodine', 'Chlorhexidine', 'Hydrogen Peroxide']
}

print("\n--- Pharmacy Inventory Report (Long Names) ---")

for med_list in pharmacy_inventory.values():
    for med in med_list:
        if len(med) > 10:
            print(f"Long name alert: {med} ({len(med)} characters).")
```
